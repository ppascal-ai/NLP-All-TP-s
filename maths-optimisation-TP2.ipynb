```python
# Import numpy and pyplot libraries
import numpy as np
import math 
import matplotlib.pyplot as plt
```

## Exercise 1 

```python
def f1(x, y):
    return x**3 + y**3 - 3*x - 12*y

def f2(x, y):
    return x**3 - 12*x*y + 8*y**3

def f3(x, y):
    return (x - 1)**2 + 10*(x**2 - y)**2
```

To find the critical points, I just set all the partial derivatives to zero and solved the system.  
Then I used the Hessian matrix to check if each point is a minimum, maximum, or saddle point.  
Itâ€™s basically just checking the sign of the eigenvalues: both positive = minimum, both negative = maximum, mixed = saddle.

### 3.1

```python
# Definition of the plotting domain
xmin, xmax, nx = -4, 4, 41
ymin, ymax, ny = -4, 4, 41
# Discretization of the plotting domain
x1d = np.linspace(xmin,xmax,nx)
y1d = np.linspace(ymin,ymax,ny)
x2d, y2d = np.meshgrid(x1d, y1d)

# Plot contour lines (isovalues) of f1
nIso = 21
plt.contour(x2d,y2d,f1(x2d,y2d),nIso)
plt.title('Contour Lines')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid()
plt.axis('square')
```

Contour lines help us visualize the function in 2D.  
Closed curves usually indicate a minimum or maximum, and curves that cross each other strangely often indicate saddle points.  
It makes the landscape of the function easier to understand.

### 3.2

```python
critical_points_f1 = [(-1, -2), (-1, 2), (1, -2), (1, 2)]

plt.contour(x2d, y2d, f1(x2d, y2d), nIso)
for (xc, yc) in critical_points_f1:
    plt.plot(xc, yc, 'ro')
    plt.text(xc + 0.1, yc + 0.1, f'({xc},{yc})')

plt.title('Contour lines of f1 with critical points')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid()
plt.axis('square')
plt.show()
```

### 3.3

```python
def df1(x, y):
    # âˆ‚f1/âˆ‚x and âˆ‚f1/âˆ‚y
    df_dx = 3*x**2 - 3
    df_dy = 3*y**2 - 12
    return [df_dx, df_dy]

# Check that the gradient is zero at the critical points
for (xc, yc) in critical_points_f1:
    grad = df1(xc, yc)
    print(f"âˆ‡f1({xc},{yc}) = {grad}")
```

### 3.4

```python
def d2f1(x, y):
    # Second derivatives
    d2f_dxdx = 6*x
    d2f_dxdy = 0
    d2f_dydx = 0
    d2f_dydy = 6*y
    return [[d2f_dxdx, d2f_dxdy],
            [d2f_dydx, d2f_dydy]]

for (xc, yc) in critical_points_f1:
    H = d2f1(xc, yc)
    eigvals = np.linalg.eigvals(H)
    print(f"Hessian at ({xc},{yc}) = {H}")
    print(f"Eigenvalues: {eigvals}\n")
```
